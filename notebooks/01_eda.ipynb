# Phase 3 EDA
Concise exploratory checks for demand forecasting using cleaned canonical data.

```python
import matplotlib.pyplot as plt
import pandas as pd

from retail_forecasting.paths import build_project_paths
```

```python
paths = build_project_paths()
cleaned_path = paths.data_interim_dir / "cleaned_retail.parquet"
df = pd.read_parquet(cleaned_path)

print(f"Loaded: {cleaned_path}")
print(f"Shape: {df.shape}")
df.head()
```

```python
print("Date coverage")
print(df['date'].min(), "->", df['date'].max())
print("Distinct days:", df['date'].nunique())
print("Stores:", df['store_id'].nunique())
print("Products:", df['product_id'].nunique())
```

```python
zero_demand_rate = float((df['units_sold'] == 0).mean())
print(f"Zero-demand rate: {zero_demand_rate:.4f}")

ax = df['units_sold'].plot(kind='hist', bins=30, title='Units Sold Distribution')
ax.set_xlabel('units_sold')
plt.show()
```

```python
store_counts = df.groupby('store_id', dropna=False).size().sort_values(ascending=False)
product_counts = df.groupby('product_id', dropna=False).size().sort_values(ascending=False)

print('Rows per store (top 10)')
print(store_counts.head(10))
print('\nRows per product (top 10)')
print(product_counts.head(10))
```

```python
ax = df['price'].plot(kind='hist', bins=30, title='Price Distribution')
ax.set_xlabel('price')
plt.show()
```

```python
if 'inventory_level' in df.columns:
    print(df['inventory_level'].describe())
    ax = df['inventory_level'].plot(kind='hist', bins=30, title='Inventory Level Distribution')
    ax.set_xlabel('inventory_level')
    plt.show()
else:
    print('inventory_level not present in cleaned data')
```

```python
if 'discount' in df.columns:
    print(df['discount'].describe())
    ax = df['discount'].plot(kind='hist', bins=30, title='Discount Distribution')
    ax.set_xlabel('discount')
    plt.show()
else:
    print('discount not present in cleaned data')
```
