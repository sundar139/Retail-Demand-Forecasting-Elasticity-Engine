# Phase 3 Manual Feature Engineering
Demonstrates leakage-safe manual feature generation via package APIs.

```python
import pandas as pd

from retail_forecasting.features_manual import (
    build_manual_features_pipeline,
    generate_manual_features_frame,
    load_cleaned_retail_data,
    load_features_summary,
)
from retail_forecasting.paths import build_project_paths
```

```python
source_path, cleaned_df = load_cleaned_retail_data()
print(f"Loaded cleaned data: {source_path}")
print(f"Shape: {cleaned_df.shape}")
cleaned_df.head()
```

```python
feature_df, feature_metadata = generate_manual_features_frame(cleaned_df, drop_warmup_rows=True)
print('Feature frame shape:', feature_df.shape)
print('Feature count:', feature_metadata['feature_column_count'])
feature_df.head()
```

```python
columns_to_inspect = [
    'date',
    'store_id',
    'product_id',
    'units_sold',
    'units_sold_lag_1',
    'units_sold_lag_7',
    'units_sold_roll_mean_7',
    'price',
    'price_lag_1',
    'price_roll_mean_7',
    'price_momentum_7',
]
available_columns = [column for column in columns_to_inspect if column in feature_df.columns]
feature_df[available_columns].head(20)
```

```python
artifacts = build_manual_features_pipeline()
for name, path in artifacts.items():
    print(name, '->', path)
```

```python
summary = load_features_summary()
summary
```

```python
paths = build_project_paths()
features_path = paths.data_processed_dir / 'features_manual.parquet'
saved_features = pd.read_parquet(features_path)
print('Saved features shape:', saved_features.shape)
saved_features.head()
```
